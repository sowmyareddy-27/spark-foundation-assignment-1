```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r"https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data["Hours"].value_counts().plot(kind="bar")
plt.show()
```

```python
data["Scores"].value_counts().plot(kind="bar")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(19,7))
sns.barplot(x = "Hours", y ="Scores", data =data)
plt.xticks(rotation=90)
plt.title("Hours vs Scores")
plt.show()
```

```python
data.plot(x='Hours',y='Scores',style='*')
plt.title('hours vs Scores')
plt.xlabel('hours')
plt.ylabel('Scores')
plt.show()
```

```python
X=data.iloc[:,0:1]
print(X)
```

```python
y=data.iloc[:,1:]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X.values, y, test_size = 0.2, random_state = 0)
print((X_train.shape),(y_train.shape))
print((X_test.shape),(y_test.shape))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
lr=LinearRegression()
lr.fit(X_train,y_train)
lr_pred=lr.predict(X_test)
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

print(mean_squared_error(lr_pred,y_test))
```

```python
print(np.sqrt(mean_squared_error(lr_pred,y_test)))
```

```python
line = lr.coef_*X+lr.intercept_

plt.scatter(X,y)
plt.plot(X,line)
plt.show()
```
